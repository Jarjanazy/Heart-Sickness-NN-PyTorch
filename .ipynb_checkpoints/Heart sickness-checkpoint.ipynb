```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.fullyConnectedLayer1 = nn.Linear(13, 64) # the number of features in the input is 13
        self.fullyConnectedLayer2 = nn.Linear(64, 64)
        self.fullyConnectedLayer3 = nn.Linear(64, 64)
        self.fullyConnectedLayer4 = nn.Linear(64, 2)
    
    def forward(self, x):
        # get the putput of the layer and pass it to the rectified linear activation function
        x = F.relu(self.fullyConnectedLayer1(x)) 
        x = F.relu(self.fullyConnectedLayer2(x))
        x = F.relu(self.fullyConnectedLayer3(x))
        # get the output of the last layer
        x = self.fullyConnectedLayer4(x)
        # feed the output to the softmax activation function
        return F.log_softmax(x, dim=1) # for the dim explantion check https://discuss.pytorch.org/t/dimension-for-logsoftmax/49042
        
```

```python
seperateLastElement = lambda arr : [arr[:-1], arr[-1]]
inputDataSeperated = list(map(seperateLastElement , df.to_numpy()))
```

```python
X = torch.FloatTensor([data[0] for data in inputDataSeperated])
y = torch.FloatTensor([data[1] for data in inputDataSeperated])
```

```python
net = Net()
```

This feeds the full data set to the network for a single feed forward pass and gives us the outputs for each data item. We notice each input has two outputs corresponding to the two calssification classes we have
